```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# E_policarbonato = 2.2 GPa
# Link: https://designerdata.nl/materials/plastics/thermo-plastics/polycarbonate
exx = np.loadtxt('../data/DatosG3/exx.txt')
exy = np.loadtxt('../data/DatosG3/exy.txt')
eyy = np.loadtxt('../data/DatosG3/eyy.txt')
```

```python
exx.shape
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
fxx = ax.imshow(exx,cmap="jet")
fig.colorbar(fxx)
ax.set_title('Deformación unitaria e_xx')
ax.set_xlabel('pixel')
ax.set_ylabel('pixel')
fig.savefig('../fig/exx.png')
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
fxy = ax.imshow(exy,cmap="jet")
fig.colorbar(fxy)
ax.set_title('Deformación unitaria e_xy')
ax.set_xlabel('pixel')
ax.set_ylabel('pixel')
fig.savefig('../fig/exy.png')
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
fyy = ax.imshow(eyy,cmap="jet")
fig.colorbar(fyy)
ax.set_title('Deformación unitaria e_yy')
ax.set_xlabel('pixel')
ax.set_ylabel('pixel')
fig.savefig('../fig/eyy.png')
```
