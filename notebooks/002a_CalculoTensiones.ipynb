```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# E_policarbonato = 2.2 GPa
# Link: https://designerdata.nl/materials/plastics/thermo-plastics/polycarbonate
exx = np.loadtxt('../data/DatosG3/exx.txt')
exy = np.loadtxt('../data/DatosG3/exy.txt')
eyy = np.loadtxt('../data/DatosG3/eyy.txt')
```

```python
# E=2.2 GPa
# Con lo cual s_ij = [MPa]
sxx = exx*2.2
sxy = exy*2.2
syy = eyy*2.2
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
ax.imshow(sxx,cmap="jet")
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
ax.imshow(sxy,cmap="jet")
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
ax.imshow(syy,cmap="jet")
```
