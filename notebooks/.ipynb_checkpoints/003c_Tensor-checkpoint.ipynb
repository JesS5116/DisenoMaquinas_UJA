```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# E_policarbonato = 2.2 GPa
# Link: https://designerdata.nl/materials/plastics/thermo-plastics/polycarbonate
u = np.loadtxt('../data/DatosG3/u.txt')
v = np.loadtxt('../data/DatosG3/v.txt')

exx = np.loadtxt('../data/DatosG3/exx.txt')
eyy = np.loadtxt('../data/DatosG3/eyy.txt')
txy = np.loadtxt('../data/DatosG3/exy.txt')
```

```python
ug = np.zeros((92,110))
#h = 0.00001
for i in np.arange(0,91,1):
    for j in np.arange(0,109,1):
        D1 = ((u[i+1][j]) - u[i][j])
        D2 = ((u[i][j+1]) - u[i][j])/1.85
        ug[i][j] = D2
        
vg = np.zeros((92,110))
for i in np.arange(0,91,1):
    for j in np.arange(0,109,1):
        D1 = ((v[i+1][j]) - v[i][j])
        D2 = ((v[i][j+1]) - v[i][j])/1.85
        vg[i][j] = D2
        
tg = np.zeros((92,110))
for i in np.arange(0,91,1):
    for j in np.arange(0,109,1):
        D1 = ((v[i+1][j]) - v[i][j])
        D2 = 0.5* (((v[i][j+1]) - v[i][j])/1.85 + ((u[i][j+1]) - u[i][j])/1.85)
        tg[i][j] = D2
```

```python
ug.shape
```

### Gráficos comparativos con Tensores e deformación

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,8))

fu = ax[0].imshow(exx,cmap="jet")
fig.colorbar(fu,fraction=0.046, pad=0.04)
ax[0].set_title('exx')
ax[0].set_xlabel('pixel')
ax[0].set_ylabel('pixel')

fug = ax[1].imshow(ug,cmap="jet")
#fig.colorbar(fug,fraction=0.046, pad=0.04)
ax[1].set_title('ug')
ax[1].set_xlabel('pixel')
ax[1].set_ylabel('pixel')
#fig.savefig('../fig/exx.png')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,8))

fv = ax[0].imshow(eyy,cmap="jet")
fig.colorbar(fv,fraction=0.046, pad=0.04)
ax[0].set_title('eyy')
ax[0].set_xlabel('pixel')
ax[0].set_ylabel('pixel')

fvg = ax[1].imshow(vg,cmap="jet")
#fig.colorbar(fug,fraction=0.046, pad=0.04)
ax[1].set_title('vg')
ax[1].set_xlabel('pixel')
ax[1].set_ylabel('pixel')
#fig.savefig('../fig/exx.png')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,8))

ft = ax[0].imshow(txy,cmap="jet")
fig.colorbar(ft,fraction=0.046, pad=0.04)
ax[0].set_title('txy')
ax[0].set_xlabel('pixel')
ax[0].set_ylabel('pixel')

ftg = ax[1].imshow(tg,cmap="jet")
#fig.colorbar(fug,fraction=0.046, pad=0.04)
ax[1].set_title('tg')
ax[1].set_xlabel('pixel')
ax[1].set_ylabel('pixel')
#fig.savefig('../fig/exx.png')
```
