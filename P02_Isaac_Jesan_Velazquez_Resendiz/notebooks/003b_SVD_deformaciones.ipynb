```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# E_policarbonato = 2.2 GPa
# Link: https://designerdata.nl/materials/plastics/thermo-plastics/polycarbonate
exx = np.loadtxt('../data/DatosG3/exx.txt')
exy = np.loadtxt('../data/DatosG3/exy.txt')
eyy = np.loadtxt('../data/DatosG3/eyy.txt')
```

```python
# Realización de Singular Value Decomposition
Uxx, Sxx, VTxx = np.linalg.svd(exx,full_matrices=False)
Sxx = np.diag(Sxx)

Uyy, Syy, VTyy = np.linalg.svd(eyy,full_matrices=False)
Syy = np.diag(Syy)

Uxy, Sxy, VTxy = np.linalg.svd(exy,full_matrices=False)
Sxy = np.diag(Sxy)
```

### Gráficos comparativos con SVD

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,8))

# ax.imshow(memo)
fxx = ax[0].imshow(exx,cmap="jet")
#fig.colorbar(fxx)
ax[0].set_title('Deformación unitaria e_xx')
ax[0].set_xlabel('pixel')
ax[0].set_ylabel('pixel')

r = 50
Xapprox_xx = Uxx[:,:r] @ Sxx[:r,:r] @ VTxx[:r,:]
img = ax[1].imshow(Xapprox_xx,cmap='jet')
ax[1].set_title(f'e_xx con SVD y r={r}')
ax[1].set_xlabel('pixel')
ax[1].set_ylabel('pixel')
fig.colorbar(img,fraction=0.046, pad=0.04)
#fig.savefig('../fig/exx.png')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,8))

# ax.imshow(memo)
fyy = ax[0].imshow(eyy,cmap="jet")
#fig.colorbar(fxx)
ax[0].set_title('Deformación unitaria e_xx')
ax[0].set_xlabel('pixel')
ax[0].set_ylabel('pixel')

r = 50
Xapprox_yy = Uyy[:,:r] @ Syy[:r,:r] @ VTyy[:r,:]
img = ax[1].imshow(Xapprox_yy,cmap='jet')
ax[1].set_title(f'e_yy con SVD y r={r}')
ax[1].set_xlabel('pixel')
ax[1].set_ylabel('pixel')
fig.colorbar(img,fraction=0.046, pad=0.04)
#fig.savefig('../fig/exy.png')
```
