```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# E_policarbonato = 2.2 GPa
# Link: https://designerdata.nl/materials/plastics/thermo-plastics/polycarbonate
exx = np.loadtxt('../data/DatosG3/exx.txt')
exy = np.loadtxt('../data/DatosG3/exy.txt')
eyy = np.loadtxt('../data/DatosG3/eyy.txt')
```

```python
# E=2.2 GPa
# Con lo cual s_ij = [MPa]
sxx = exx*2.2
sxy = exy*2.2
syy = eyy*2.2
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
fxx = ax.imshow(sxx,cmap="jet")
fig.colorbar(fxx)
ax.set_title('Mapa de esfuerzos s_xx')
ax.set_xlabel('pixel')
ax.set_ylabel('pixel')
fig.savefig('../fig/sxx.png')
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
fxy = ax.imshow(sxy,cmap="jet")
fig.colorbar(fxy)
ax.set_title('Mapa de esfuerzos s_xy')
ax.set_xlabel('pixel')
ax.set_ylabel('pixel')
fig.savefig('../fig/sxy.png')
```

```python
fig, ax = plt.subplots()

# ax.imshow(memo)
fyy = ax.imshow(syy,cmap="jet")
fig.colorbar(fyy)
ax.set_title('Mapa de esfuerzos s_yy')
ax.set_xlabel('pixel')
ax.set_ylabel('pixel')
fig.savefig('../fig/syy.png')
```
